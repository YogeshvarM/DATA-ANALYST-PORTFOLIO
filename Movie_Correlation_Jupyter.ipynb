```python
#Lets convert the object type to Category for finding corelation between them and mapping the categorical data
#Genrally the correlation can be done for numerical values not for object or other types
```

```python
df_conversion=df
df_conversion.dtypes
```

```python
#Conversion

for col in df_conversion.columns:
    if(df_conversion[col].dtype=='object'):
        df_conversion[col]=df_conversion[col].astype('category')
        df_conversion[col]=df_conversion[col].cat.codes #mapping 
```

```python
#after conversion 

df_conversion.dtypes
```

```python
df_conversion.head()
```

```python
corelation_matrix=df_conversion.corr()
sns.heatmap(corelation_matrix,annot=True)
plt.title('Corelation between features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corelation_matrix=df_conversion.corr(method='kendall')
sns.heatmap(corelation_matrix,annot=True)
plt.title('Corelation between features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corelation_matrix=df_conversion.corr(method='spearman')
sns.heatmap(corelation_matrix,annot=True)
plt.title('Corelation between features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#unstacking
corelation_matrix_conversion=df_conversion.corr()
pairing=corelation_matrix_conversion.unstack() #Unstack -Pivot a level of the (necessarily hierarchical) index labels, returning a DataFrame having a new level of column labels whose inner-most level consists of the pivoted index labels.
pairing.head()
```

```python
#Sorting 
#Pandas sort_values() function sorts a data frame in Ascending or Descending order of passed Column. Its different than the sorted Python function since it cannot sort a data frame and particular column cannot be selected.
sorting_pairs=pairing.sort_values()
sorting_pairs.head()
```

```python
#To_find_strong_strongest_corelation_coefficient
strong_pairs=sorting_pairs[abs(sorting_pairs)>0.5]
strong_pairs
```
